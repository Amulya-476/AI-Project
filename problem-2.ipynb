```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import numpy as np

# Load the Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

print("First few rows of the Iris dataset:")
print(df.head())

# 1. Bar chart: Count of each species
species_count = df['species'].value_counts()
species_count.plot(kind='bar', color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Bar Chart - Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 2. Line graph: Mean of features per species
df_grouped = df.groupby('species').mean().T
for species in df_grouped.columns:
    plt.plot(df_grouped.index, df_grouped[species], label=species)
plt.title('Line Graph - Mean of Features per Species')
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.legend()
plt.tight_layout()
plt.show()

# 3. Pie chart: Distribution of species
species_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Pie Chart - Distribution of Iris Species')
plt.ylabel('')
plt.tight_layout()
plt.show()

# 4. Scatter plot: Sepal length vs Sepal width
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.title('Scatter Plot - Sepal Length vs Sepal Width')
plt.tight_layout()
plt.show()

# 5. Histogram: Petal length distribution
sns.histplot(data=df, x='petal length (cm)', hue='species', multiple='stack')
plt.title('Histogram - Petal Length Distribution')
plt.tight_layout()
plt.show()

# 6. Heatmap: Correlation matrix
corr = df.iloc[:, :-1].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap - Feature Correlation')
plt.tight_layout()
plt.show()

# 7. Box plot: Sepal length per species
sns.boxplot(data=df, x='species', y='sepal length (cm)')
plt.title('Box Plot - Sepal Length per Species')
plt.tight_layout()
plt.show()

print("All graphs have been displayed successfully.")
```
